```python
import matplotlib.pyplot as plt

def plot_result(statNms,legendArr, legendNm,data_type, ylabel, number=0):
    
    if ylabel == 'Cross Entropy':
        tail = '_ce'
    elif ylabel == 'Accuracy':
        tail = '_acc'
    else:
        tail = '_' + ylabel
   
    #plt.figure(figsize=(6.40, 6.80), dpi=100)
    
    plt.figure(number)
    plt.clf()
    cnt = 0
    for statNm in statNms:
        dataArr = np.array(stats[statNm][data_type + tail])
        plt.plot(dataArr[:, 0], dataArr[:, 1], label=legendNm + " " +legendArr[cnt])
        cnt+=1
    plt.xlabel('Epoch')
    plt.ylabel(data_type+ " " +ylabel)
    plt.legend()
    plt.draw()

    
    
momentumArr = ["0.0","0.5","0.9"]
lrArr = ["0.100","0.010","0.001"]
miniBatchArr = ['10','100','500']
cnt = 0
plot_result(['MNN_lr0.010_momentum'+i+'_minibatch100_epoch1000' for i in momentumArr ],momentumArr,"momentum","train","Cross Entropy",cnt)
cnt += 1
plot_result(['MNN_lr0.010_momentum'+i+'_minibatch100_epoch1000' for i in momentumArr ],momentumArr,"momentum","train","Accuracy",cnt)
cnt += 1
plot_result(['MNN_lr0.010_momentum'+i+'_minibatch100_epoch1000' for i in momentumArr ],momentumArr,"momentum","valid","Cross Entropy",cnt)
cnt += 1
plot_result(['MNN_lr0.010_momentum'+i+'_minibatch100_epoch1000' for i in momentumArr ],momentumArr,"momentum","valid","Accuracy",cnt)
cnt += 1

plot_result(['MNN_lr'+i+'_momentum0.9_minibatch100_epoch1000' for i in lrArr ],lrArr,"lr","train","Cross Entropy",cnt)
cnt += 1
plot_result(['MNN_lr'+i+'_momentum0.9_minibatch100_epoch1000' for i in lrArr ],lrArr,"lr","train","Accuracy",cnt)
cnt += 1
plot_result(['MNN_lr'+i+'_momentum0.9_minibatch100_epoch1000' for i in lrArr ],lrArr,"lr","valid","Cross Entropy",cnt)
cnt += 1
plot_result(['MNN_lr'+i+'_momentum0.9_minibatch100_epoch1000' for i in lrArr ],lrArr,"lr","valid","Accuracy",cnt)
cnt += 1

plot_result(['MNN_lr0.010_momentum0.9_minibatch'+i+'_epoch1000' for i in miniBatchArr ],miniBatchArr,"batch size","train","Cross Entropy",cnt)
cnt += 1
plot_result(['MNN_lr0.010_momentum0.9_minibatch'+i+'_epoch1000' for i in miniBatchArr ],miniBatchArr,"batch size","train","Accuracy",cnt)
cnt += 1
plot_result(['MNN_lr0.010_momentum0.9_minibatch'+i+'_epoch1000' for i in miniBatchArr ],miniBatchArr,"batch size","valid","Cross Entropy",cnt)
cnt += 1
plot_result(['MNN_lr0.010_momentum0.9_minibatch'+i+'_epoch1000' for i in miniBatchArr ],miniBatchArr,"batch size","valid","Accuracy",cnt)
cnt += 1


```
